```python
# Step 1: Install Hugging Face Transformers library
!pip install transformers

# Step 2: Import necessary packages
from transformers import pipeline

# Step 3: Load a pre-trained model and tokenizer
# Here we use the 'distilbert-base-uncased' model for simplicity
nlp_pipeline = pipeline('sentiment-analysis')

# Step 4: Tokenize the input text and run the model
# Define your input text
input_text = "I love learning about natural language processing!"

# Step 5: Get predictions
result = nlp_pipeline(input_text)

# Step 6: Print the results
print(result)
```
